```python
nums = list(range(0, 1000001))
```

```python
nums_rdd = sc.parallelize(nums)
nums_rdd
```

```python
nums_rdd.take(5)
```

```python
squared_nums_rdd = nums_rdd.map(lambda x: x**2)
```

```python
squared_nums_rdd.take(5)
```

```python
len(str(546))
```

```python
length_rdd_pairs = nums_rdd.map(lambda x: (x**2, len(str(x** 2))))
```

```python
length_rdd_pairs.take(25)
```

```python
even_digit_pairs = length_rdd_pairs.map(lambda x: (x[1]%2) == 0)  #if the cube is even or odd, return boolean in a list
```

```python
even_digit_pairs.take(25)
```

Group num of digits together

```python
flipped_keys = length_rdd_pairs.map(lambda x: (x[1], x[0]))
#Make digit the key
```

```python
flipped_keys.take(10)
```

```python
#Agg by keys
group_by_keys = flipped_keys.groupByKey()
```

```python
group_by_keys.take(25)
```

```python
group_by_keys = group_by_keys.map(lambda x: (x[0], list(x[1])))
group_by_keys.take(1)
```

```python
#get the average in a group
average = group_by_keys.map(lambda x: (x[0], sum(x[1])/len(x[1])))
```

```python
average.take(5)
```
